Conduct a Univariate Analysis of the participants' Cholesterol. What is noticeable about it? Create a function that normalizes the participants' Cholesterol and create two visualizations, one before the `"Chol"` data is modified, and one after normalizing the `"Chol"` data.

```python
# YOUR CODE HERE
def chol_plot_and_norm(data):
    plt.subplot(1,2,1)
    data.hist()
    plt.plot()
    plt.title('Initial Chol')

    q3, q1 = np.percentile(data, [75 ,25])
    iqr = q3-q1
    max_range = 1.5 * iqr
    norm_data = data[(data<q3+max_range) & (data>q1-max_range)]

    plt.subplot(1,2,2)
    norm_data.hist()
    plt.plot()
    plt.title('Nomalized Chol')

chol_plot_and_norm(data.iloc[:, 11])
```

Fill in the function below. The function recieves a non-empty string and an int n, and returns a new string where the character at index n has been removed.


```python
def remove_char(old_string, n):
    # YOUR CODE HERE
    first = old_string[0:n]
    last = old_string[n+1:]
    new_string = first + last
    return new_string
```

Given a string, return a new string where the first and last characters have been exchanged.

```python
def swap_chars(old_string):
  # YOUR CODE HERE
  new_string = old_string[len(old_string)-1] + old_string[1:len(old_string)-1] + old_string[0]
  return new_string

swap_chars("jfksdjkfsdlfjlsjfkladjfkla")
```

Test the function with three different parameters


```python
# YOUR CODE HERE
remove_char("nknknefnslfknalksfn", 5)
remove_char("asdf", 1)
remove_char("asdfsafefefsasfsafsadfsa", 3)

swap_chars("jfksdjkfsdlfjlsjfkladjfkla")
swap_chars("dfjsdjfk")
swap_chars("sfjkfewnklfklw")
```
